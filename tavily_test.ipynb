### This is the notebook for testing TAVILY search engine for ai agents

NOTE: We need to store the api key with the name TAVILY_API_KEY, then langchain will fetch the key automatically ( this is from the docs )

```python
from langchain_community.tools.tavily_search import TavilySearchResults
from dotenv import load_dotenv
load_dotenv()
```

Note that this will perfomr a parallel search for all the queries that we give to it in a list.

```python
tool = TavilySearchResults(max_results=2)
results = tool.invoke({"query": 'ChatGPT version releases and updates from 2022 to 2026'})
results
```

```python
for r in results:
    print(r['content'])
```
